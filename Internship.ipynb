```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from keras import backend as K
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import glob
import os
import cv2
import sys
```

```python
df=pd.read_csv("/data (1) (1).csv")
#filtering data
df.drop(columns=['perimeter_mean','smoothness_mean','concave points_mean','perimeter_worst','smoothness_worst','concave points_worst','symmetry_worst','fractal_dimension_worst','Unnamed: 32'])
df.diagnosis = [1 if each == "M" else 0 for each in df.diagnosis]

m_df = df.loc[df['diagnosis'] == 1]

b_df = df.loc[df['diagnosis'] == 0]

b_df.head(10)
```

```python
#training and testing data

x = df.iloc[:, 1:]
y = df['diagnosis'].tolist()

# Share test and train data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
y_train = np.array(y_train)
y_test = np.array(y_test)

print(y_test.shape)

print('Train data: X=%s, y=%s' % (x_train.shape, y_train.shape))
print('Test data: X=%s, y=%s' % (x_test.shape, y_test.shape))


x_train = x_train.values.reshape((x_train.shape[0],x_train.shape[1], 1))
x_test  = x_test.values.reshape((x_test.shape[0], x_test.shape[1], 1))
print('Train data: X=%s, y=%s' % (x_train.shape, y_train.shape))
print('Test data: X=%s, y=%s' % (x_test.shape, y_test.shape))


print('x train shape : ', x_train.shape)
print('y_train shape : ', y_train.shape)
print('x_test shape : ', x_test.shape)
print('y_test shape : ', y_test.shape)


print(y_train)

print(x_train)


x_train2 = x_train/255 - 0.5
x_test2 = x_test/255 - 0.5

cancer_classes = [1,0]


y_train_indexes = []
for classs in y_train:
    y_train_indexes.append(cancer_classes.index(classs))
    

```

```python
#for tarining model

from statistics import mean
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling2D, Flatten, Dense, Activation, Dropout, BatchNormalization
from keras.layers.advanced_activations import LeakyReLU
```

```python
# model
model=Sequential()
# layers
model.add(Conv1D(filters=32, kernel_size=2, activation='relu', input_shape=(32,1)))
model.add(BatchNormalization())
model.add(Dropout(0.2))


model.add(Conv1D(filters=64, kernel_size=2, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100, validation_data=(x_test,y_test))
```
